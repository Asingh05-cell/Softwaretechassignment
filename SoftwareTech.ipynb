```python
import pandas as pd

import numpy as np

from sklearn.model_selection import KFold

from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.svm import SVR

from sklearn.neighbors import KNeighborsRegressor

from sklearn.preprocessing import StandardScaler, OneHotEncoder

from sklearn.impute import SimpleImputer

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import joblib



# Load the dataset

data = pd.read_csv('Electricity.csv', low_memory=False)



# Define target variables

target_variables = ['SMPEA', 'SMPEP2']



# Define features (drop target variables and date/time columns)

features = data.drop(target_variables + ['DateTime'], axis=1)



# Identify categorical and numeric columns

categorical_columns = ['Holiday', 'DayOfWeek', 'PeriodOfDay']

numeric_columns = ['SystemLoadEA', 'ORKTemperature', 'ORKWindspeed', 'CO2Intensity', 'SystemLoadEP2']



# Replace '?' and other non-numeric values in numeric columns with NaN

# and convert columns to float

for col in numeric_columns:

    features[col] = features[col].replace('?', np.nan).astype(float)



# Handle missing values in numeric columns using imputation (e.g., mean)

numeric_transformer = Pipeline([

    ('imputer', SimpleImputer(strategy='mean')),

    ('scaler', StandardScaler())

])



# Handle categorical columns using one-hot encoding

categorical_transformer = OneHotEncoder(handle_unknown='ignore')



# Create a column transformer that combines numeric and categorical transformers

preprocessor = ColumnTransformer(

    transformers=[

        ('num', numeric_transformer, numeric_columns),

        ('cat', categorical_transformer, categorical_columns)

    ]

)



# Create a pipeline that includes the preprocessor

pipeline = Pipeline([

    ('preprocessor', preprocessor)

])



# Apply transformations to the features

features_transformed = pipeline.fit_transform(features)



# Clean and convert target variables to numeric

data[target_variables] = data[target_variables].replace('?', np.nan).astype(float)



# Handle missing values in target variables using mean imputation

imputer = SimpleImputer(strategy='mean')

data[target_variables] = imputer.fit_transform(data[target_variables])



# Define models to evaluate

models = {

    "Linear Regression": LinearRegression(),

    "Decision Tree": DecisionTreeRegressor(max_depth=10),

    "Random Forest": RandomForestRegressor(n_estimators=100, max_depth=10, n_jobs=-1),

    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100),

    "Support Vector Regression": SVR(kernel='linear'),

    "K-Nearest Neighbors": KNeighborsRegressor(n_neighbors=5, n_jobs=-1)

}



# Initialize KFold cross-validation

kf = KFold(n_splits=5, shuffle=True, random_state=42)



# Dictionary to store evaluation metrics for each model for each target

results = {}



# Train a model for each target variable

for target in target_variables:

    results[target] = {}

    for name, model in models.items():

        fold_mae = []

        fold_mse = []

        fold_r2 = []



        # Iterate through each fold

        for train_index, test_index in kf.split(features_transformed):

            # Get training and testing data for the current fold

            X_train, X_test = features_transformed[train_index], features_transformed[test_index]

            y_train, y_test = data[target].iloc[train_index], data[target].iloc[test_index]  # Adjust to use only one target



            # Train the model on the training data

            model.fit(X_train, y_train)



            # Predict on the testing data

            y_pred = model.predict(X_test)



            # Calculate evaluation metrics

            mae = mean_absolute_error(y_test, y_pred)

            mse = mean_squared_error(y_test, y_pred)

            r2 = r2_score(y_test, y_pred)



            # Store the metrics

            fold_mae.append(mae)

            fold_mse.append(mse)

            fold_r2.append(r2)



        # Calculate average metrics across all folds

        average_mae = np.mean(fold_mae)

        average_mse = np.mean(fold_mse)

        average_r2 = np.mean(fold_r2)



        # Store results

        results[target][name] = {

            "Average MAE": average_mae,

            "Average MSE": average_mse,

            "Average R²": average_r2

        }



        # Print the results for each model

        print(f"{target} - {name}: Average MAE = {average_mae:.4f}, Average MSE = {average_mse:.4f}, Average R² = {average_r2:.4f}")



    # Compare the results and find the best model based on R² for each target

    best_r2 = -float('inf')

    best_model_name = None

    best_model = None

    for name, metrics in results[target].items():

        if metrics["Average R²"] > best_r2:

            best_r2 = metrics["Average R²"]

            best_model = models[name]

            best_model_name = name

    print(f"The best model for {target} is: {best_model_name} with R² = {best_r2:.4f}")



    # Save the best model for each target

    joblib.dump(best_model, f'best_model_{target}.joblib')

    print(f"The best model ({best_model_name}) for {target} has been saved as 'best_model_{target}.joblib'")
```

```python
#Deployment of the best model in production with web app
# for this part of the code pls put it into pycharm to see the gui.


from flask import Flask, render_template_string, request

import joblib

import pandas as pd



# Create a Flask app

app = Flask(__name__)



# Load the preprocessing pipeline and best model

preprocessing_pipeline_file = 'preprocessing_pipeline.joblib'

best_model_file = 'best_model_SMPEA.joblib'  # Adjust based on the target variable you want



# Load the files

try:

    pipeline = joblib.load(preprocessing_pipeline_file)

    best_model = joblib.load(best_model_file)

except FileNotFoundError as e:

    raise FileNotFoundError(f"Error: {e}")



# Define a route for the home page

@app.route('/', methods=['GET', 'POST'])

def home():

    # HTML form template

    html_template = '''

    <!DOCTYPE html>

    <html>

    <head>

        <title>ML Prediction Calculator</title>

    </head>

    <body>

        <h1>ML Prediction Calculator</h1>

        <form method="POST">

            <label for="system_load_ea">System Load EA:</label><br>

            <input type="number" step="0.01" id="system_load_ea" name="system_load_ea" required><br><br>



            <label for="ork_temperature">ORK Temperature:</label><br>

            <input type="number" step="0.01" id="ork_temperature" name="ork_temperature" required><br><br>



            <label for="ork_windspeed">ORK Windspeed:</label><br>

            <input type="number" step="0.01" id="ork_windspeed" name="ork_windspeed" required><br><br>



            <label for="co2_intensity">CO2 Intensity:</label><br>

            <input type="number" step="0.01" id="co2_intensity" name="co2_intensity" required><br><br>



            <label for="system_load_ep2">System Load EP2:</label><br>

            <input type="number" step="0.01" id="system_load_ep2" name="system_load_ep2" required><br><br>



            <label for="holiday">Holiday (Yes/No):</label><br>

            <input type="text" id="holiday" name="holiday" required><br><br>



            <label for="day_of_week">Day of Week:</label><br>

            <input type="text" id="day_of_week" name="day_of_week" required><br><br>



            <label for="period_of_day">Period of Day:</label><br>

            <input type="text" id="period_of_day" name="period_of_day" required><br><br>



            <input type="submit" value="Predict">

        </form>

        {% if prediction %}

            <h2>Prediction:</h2>

            <p>The predicted value is: {{ prediction }}</p>

        {% endif %}

    </body>

    </html>

    '''



    if request.method == 'POST':

        # Retrieve user inputs from the form

        system_load_ea = float(request.form.get('system_load_ea'))

        ork_temperature = float(request.form.get('ork_temperature'))

        ork_windspeed = float(request.form.get('ork_windspeed'))

        co2_intensity = float(request.form.get('co2_intensity'))

        system_load_ep2 = float(request.form.get('system_load_ep2'))

        holiday = request.form.get('holiday')

        day_of_week = request.form.get('day_of_week')

        period_of_day = request.form.get('period_of_day')



        # Create a dictionary of inputs

        new_input_data = {

            'SystemLoadEA': system_load_ea,

            'ORKTemperature': ork_temperature,

            'ORKWindspeed': ork_windspeed,

            'CO2Intensity': co2_intensity,

            'SystemLoadEP2': system_load_ep2,

            'Holiday': holiday,

            'DayOfWeek': day_of_week,

            'PeriodOfDay': period_of_day,

        }



        # Convert the inputs to a DataFrame

        new_input_df = pd.DataFrame([new_input_data])



        # Preprocess the new input data

        new_input_transformed = pipeline.transform(new_input_df)



        # Predict using the best model

        prediction = best_model.predict(new_input_transformed)



        # Display the prediction in the HTML template

        return render_template_string(html_template, prediction=prediction[0])



    # Render the HTML form

    return render_template_string(html_template)



# Run the Flask app

app.run(host='0.0.0.0', port=5000)
```
